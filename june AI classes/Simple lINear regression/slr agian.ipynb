```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading data

data = pd.read_csv("placements_new.csv")
```

```python
data
```

```python
plt.scatter(data["cgpa"], data["package"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
# x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
data.head()
```

```python
x = data.iloc[ : , 0 : 1]
x
```

```python
y = data["package"]
y
```

```python
# x = [1, 2, 3, 4, 5]

# x[0 : 2]
```

```python
# splitting data in two parts

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train.shape
```

```python
# traning model

# regressions ---> Simple Linear regression

# LinearRegression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# create instance of a class

model = LinearRegression()
```

```python
# fit()   -----> this method is use for traing the model

model.fit(x_train, y_train)
```

```python
m = model.coef_
m
```

```python
b = model.intercept_
b
```

```python
# y = mx + b

y = 0.57467027 * 9 + (-1.0076134311677745)
y
```

```python
y_pred = model.predict(x_test)
```
