```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# bar chart
```

```python
months = np.array(["Jan", "Feb", "March", "April", "May"])
profit = np.array([10, 6, 7, 9, 6])
```

```python
# plt.bar(x_axis, y_axis)

plt.bar(months, profit)
```

```python
plt.bar(months, profit)
plt.xlabel("Months for profit")
plt.ylabel("Profit per month")
plt.title("My Revenue")
```

```python
plt.bar(months, profit)
plt.xlabel("Months for profit")
plt.ylabel("Profit per month")
plt.title("My Revenue")
# plt.grid()
```

```python
# Chnage color of bars
colorss = ["red", "green", "yellow", "lightgreen", "blue"]

plt.bar(months, profit, color = colorss)
plt.xlabel("Months for profit")
plt.ylabel("Profit per month")
plt.title("My Revenue")
```

```python

# Chnage color of bars
colorss = ["red", "green", "yellow", "lightgreen", "blue"]

plt.bar(months, profit, color = colorss, label = "Profit Bars")
plt.xlabel("Months for profit")
plt.ylabel("Profit per month")
plt.title("My Revenue")

plt.plot(months, profit, color = "brown", linestyle = "dashdot", label = "Profit Trends")

plt.legend()

# months = np.array(["Jan", "Feb", "March", "April", "May"])
# profit = np.array([10, 6, 7, 9, 6])

                    # 5
for i in range(len(months)):
                        # ("jan", 10)
    plt.annotate(xy = (months[i], profit[i]), text = f"{profit[i]} Lakh")
```
