```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12, 34, 45, 67, 78, 100])
y = np.array([34, 56, 67, 89, 90, 93])
```

```python
# plot
```

```python
plt.plot(x, y, marker = "*", ms = 15, mec = "red", mfc = "yellow")

for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})")

plt.grid()
```

```python
plt.plot(x, y, marker = ".", ms = 15, mec = "red", mfc = "yellow")

for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})")

plt.grid(axis = "both")

# 'x'
# 'y'
# 'both'
```

```python
plt.plot(x, y, marker = ".", ms = 15, mec = "red", mfc = "yellow")

for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})")

plt.grid(axis = "both")
plt.xlabel("Age")
plt.ylabel("Salllery")
plt.title("this is info", color = "red")
# 'x'
# 'y'
# 'both'
```

```python
plt.plot(x, y, marker = ".", ms = 15, mec = "red", mfc = "yellow", color = "blue", linestyle = "dashdot")

# 'dashed'
# 'dotted'
# 'dashdot'
# 'solid'

for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})")

plt.grid(axis = "both")
plt.xlabel("Age")
plt.ylabel("Salllery")
plt.title("this is info", color = "red")
# 'x'
# 'y'
# 'both'
```

```python
x = np.array([12, 34, 56, 78, 90, 93])
y = np.array([12, 34, 56, 78, 56, 78])

a = np.array([34, 56, 67, 89, 78])
b = np.array([23, 45, 56, 34, 23])
```

```python
plt.plot(x, y, marker = "*", ms = 12, mfc = "yellow", mec = "blue", label = "Company Info")
plt.plot(a, b, marker = ".", ms = 13, mfc = "pink", mec = "red", label = "Employee info")

plt.legend()
plt.xlabel("Salleries")
plt.ylabel("Ages")
plt.title("this is Orgnization")
plt.grid()

for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})", color = "orange")

for i in range(len(a)):
    plt.text(a[i], b[i], f"({a[i]}, {b[i]})", color = "red")
```
