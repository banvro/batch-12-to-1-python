```python
x = np.array(["Revnew", "Profit", "Lose", "Totle"])
y = np.array([400, 300, 100, 500])
```

```python
mycolor = ["green", "blue", "red", "gray"]
```

```python
plt.bar(x, y, color = mycolor)

for i in range(len(x)):
    plt.text(x[i], y[i], f"({y[i]})", color = "red")
```

```python
mylg = ["Revnew", "Profit", "Lose", "Totle"]
```

```python
plt.bar(x, y, color = mycolor, label = mylg)

plt.legend()

for i in range(len(x)):
    plt.text(x[i], y[i], f"({y[i]})", color = "red")
```

```python
plt.bar(x, y, color = mycolor, label = mylg)

plt.legend()

for i in range(len(x)):
    plt.text(x[i], y[i], f"({y[i]})", color = "red")

plt.xlabel("My Lables")
plt.ylabel("Values")
plt.title("Company Info", color = "red")

plt.grid(linestyle = "dotted", color = "red")
```
