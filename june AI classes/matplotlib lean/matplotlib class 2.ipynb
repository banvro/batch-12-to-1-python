```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
userage = np.array([20, 30, 34, 40, 50, 70])
salleries = np.array([10, 30, 20, 40, 50, 42])
```

```python
plt.plot(userage, salleries)
```

```python
plt.plot(userage, salleries, marker = ".", ms = 12)
```

```python
plt.plot(userage, salleries, marker = ".", ms = 12, mec = "red", mfc = "yellow")
```

```python
plt.plot(userage, salleries, 
         marker = ".", 
         ms = 12, 
         mec = "red", 
         mfc = "yellow",
         color = "green", 
         linewidth = 2
        )
```

```python
plt.plot(userage, salleries, 
         marker = ".", 
         ms = 12, 
         mec = "red", 
         mfc = "yellow",
         color = "green", 
         linewidth = 2,
         linestyle = "dashdot"
        )
```

```python
plt.plot(userage, salleries, 
         marker = ".", 
         ms = 12, 
         mec = "red", 
         mfc = "yellow",
         color = "green", 
         linewidth = 2,
         linestyle = "dashdot"
        )
plt.xlabel("Emp. Age")
plt.ylabel("Emp. Sallerys")
plt.title("Employee Info", loc = "left", color = "red")
```

```python
plt.plot(userage, salleries, 
         marker = ".", 
         ms = 12, 
         mec = "red", 
         mfc = "yellow",
         color = "green", 
         linewidth = 2,
         linestyle = "dashdot"
        )
plt.xlabel("Emp. Age")
plt.ylabel("Emp. Sallerys")
plt.title("Employee Info", loc = "left", color = "red")
plt.grid(color = "lightgreen")
```

```python
plt.plot(userage, salleries, 
         marker = ".", 
         ms = 12, 
         mec = "red", 
         mfc = "yellow",
         color = "green", 
         linewidth = 2,
         linestyle = "dashdot"
        )
plt.xlabel("Emp. Age")
plt.ylabel("Emp. Sallerys")
plt.title("Employee Info", loc = "left", color = "red")
plt.grid(color = "lightgreen")

# plt.annotate(xy = (50, 15), text = "hellooo")

for i in range(len(userage)):
    plt.annotate(xy = (userage[i], salleries[i]), text = f"({userage[i]}, {salleries[i]})")
```

```python
motth = np.array([2, 4, 5, 6, 3, 5, 3, 2, 4, 5])
sales = np.array([200, 400, 200, 300, 455, 344, 666, 233, 100, 400])

prfot = [200, 500, 300, 400, 300, 600, 120, 488, 500, 200]

# plt.annotate(xy = (3, 100), text="noooo")

coor = ["red", "green", "yellow", "pink", "blue", "orange", "gray", "lightgreen", "brown", "black"]

for i in range(len(motth)):
    plt.annotate(xy = (motth[i], sales[i]), text=f"({motth[i]}, {sales[i]})")
```

```python
plt.scatter(motth, sales, sizes = prfot, color = coor)

for i in range(len(motth)):
    plt.annotate(xy = (motth[i], sales[i]), text=f"({motth[i]}, {sales[i]})")

plt.grid()
```
