```python
x = np.array([10, 20, 40, 50, 70])
y = np.array([1, 4, 2, 4, 5]) 
```

```python
# plt.plot(x_axis, y_axis)

plt.plot(x, y)
```

```python
# marker

plt.plot(x, y, marker = "*")
```

```python
# ms =marker size

plt.plot(x, y, marker = "*", ms = 20)
```

```python
# marker color:

# mec => marker edge color
# mfc => marker face color


plt.plot(x, y, marker = "*", ms = 20, mec = "red")
```

```python
# marker color:

# mec => marker edge color
# mfc => marker face color


plt.plot(x, y, marker = "*", ms = 20, mec = "red", mfc = "yellow")
```

```python
# marker color:

# mec => marker edge color
# mfc => marker face color

# 1, 2, *, ., ^, >, <, x, o, 

plt.plot(x, y, marker = "o", ms = 20, mec = "red", mfc = "yellow")
```

```python
# marker color:

# line color

plt.plot(x, y, marker = "*", ms = 15, mec = "red", mfc = "yellow", color = "green")
```

```python
# line styles


plt.plot(x, y, marker = "*", ms = 15, mec = "red", mfc = "yellow", color = "green", linewidth = 3, linestyle = "dashed")
```

```python
# line styles


plt.plot(x, y, marker = "*", ms = 15, mec = "red", mfc = "yellow", color = "green", linewidth = 3, linestyle = "dotted")
```

```python
# line styles


plt.plot(x, y, marker = "*", ms = 15, mec = "red", mfc = "yellow", color = "green", linewidth = 3, linestyle = "dashdot")
```

```python
# line styles


plt.plot(x, y, marker = "*", ms = 15, mec = "red", mfc = "yellow", color = "green", linewidth = 3, linestyle = "--")
```

```python
# All line styles:

# '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

plt.plot(x, y, marker = "*", ms = 15, mec = "red", mfc = "yellow", color = "green", linewidth = 3, linestyle = ":")
```

```python
months = np.array([1, 2, 3, 4, 5, 6, 7])

profit = np.array([20, 10, 15, 20, 34, 23, 20])
```

```python
plt.plot(months, profit, marker = ".", ms = 10)
plt.xlabel("All Months")
plt.ylabel("My Profit Per Month")
plt.title("This Year Revnue", color = "red")
```

```python
# apply grid

plt.plot(months, profit, marker = ".", ms = 10)
plt.xlabel("All Months")
plt.ylabel("My Profit Per Month")
plt.title("This Year Revnue", color = "red")
plt.grid()
```

```python
# apply grid

plt.plot(months, profit, marker = ".", ms = 10)
plt.xlabel("All Months")
plt.ylabel("My Profit Per Month")
plt.title("This Year Revnue", color = "red")
plt.grid(color = "yellow")
```

```python
# apply grid

plt.plot(months, profit, marker = ".", ms = 10)
plt.xlabel("All Months")
plt.ylabel("My Profit Per Month")
plt.title("This Year Revnue", color = "red")
plt.grid(color = "green", linestyle = "-.")
```
