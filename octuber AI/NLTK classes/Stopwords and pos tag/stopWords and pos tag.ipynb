```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    xyz = [word for word in words if word.lower() not in stopwords.words('english')]
    sentences[i] = ' '.join(xyz)
    
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
stemm = WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    xyz = [stemm.lemmatize(word, pos = 'v') for word in words if word.lower() not in stopwords.words('english')]
    sentences[i] = ' '.join(xyz)
```
