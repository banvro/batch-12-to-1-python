```python
import numpy as np
```

```python
x = 6 * np.random.rand(200, 1) - 3
y = 0.8 * x ** 2 + 0.9 * x + 2 + np.random.randn(200, 1)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y, color = "red")
plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.title("Genrated Record")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train.shape
```

```python
x.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
obj = LinearRegression()
```

```python
obj.fit(x_train, y_train)
```

```python
y_pred = obj.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
plt.scatter(x, y)
plt.plot(x_test, obj.predict(x_test), color = "red")
```

## Apply Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
ploy = PolynomialFeatures(degree = 2)
```

```python
x_train_new = ploy.fit_transform(x_train)
```

```python
x_train
```

```python
x_train_new
```

```python
x_test_new = ploy.transform(x_test)
```

```python
x_test_new
```

```python
new_obj = LinearRegression()
```

```python
new_obj.fit(x_train_new, y_train)
```

```python
new_obj.coef_
```

```python
new_obj.intercept_
```

```python
x_new = np.linspace(-3, 3, 200).reshape(200, 1)
x_new_ploy = ploy.transform(x_new)
y_new = new_obj.predict(x_new_ploy)
```

```python
plt.scatter(x, y, label = "Data set")
plt.plot(x_new, y_new, color = "red", linewidth = 3, label = "Predicted Line")
plt.legend()
```
